```python
class perso:    #(tk.tk)pour quele herite des attribut peux etre
    def __init__(self,m, a, c):
        self.name = m
        self.xp = 0
        self.life = 100
        self.arme = a
        self.credit = c


    def attacked(self, nv):
        if nv == 1 :
            self.life == self.life - 5
        elif nv == 2 :
            self.life == self.life -10
        elif nv == 3:
            self.life == self.life -20
    
    


    def potion(self, nv): 
        if nv == 1 :
            self.life == self.life + 5
        elif nv == 2 :
            self.life == self.life + 10
        elif nv == 3:
            self.life == self.life + 20
    
    def XP(self, x): 
        if x == 1 :
            self.xp == self.xp + 5
        elif x == 2 :
            self.xp == self.xp + 10
        elif x == 3:
            self.xp == self.xp + 20
```

```python

class Inventory:
    def __init__(self, master):
        self.master = master
        
        master.title("Inventory")
        
        master.attributes("-topmost", True)
        
        scrollbar = tk.Scrollbar(master)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
    
        self.listbox = tk.Listbox(master, yscrollcommand=scrollbar.set)
        self.listbox.config(bg='#FAFAEB')
        self.listbox.pack(side=tk.LEFT, fill=tk.BOTH)
        scrollbar.config(command=self.listbox.yview)
        
        
    

        



        self.listbox.bind("<<ListboxSelect>>", self.remove_selected)

        self.toggle_button = tk.Button(master, text="Inventaire", command=self.toggle_listbox, activebackground='#78d6ff')
        self.toggle_button.pack()
        self.toggle_button.config(bg='blue')
        self.toggle_button.after(5000, lambda: self.toggle_button.config(bg='green'))
        self.toggle_button.config(bg='red')
        self.toggle_button.after(5000, lambda: self.toggle_button.config(bg='yellow'))

    def toggle_listbox(self):
        if self.listbox.winfo_ismapped():
            self.listbox.pack_forget()
        else:
            self.listbox.pack()
            
    def remove_selected(self, event):
        selected_item = self.listbox.curselection()
        if selected_item:
            self.listbox.delete(selected_item)

    def add_item(self):
        item = self.add_item_entry.get()
        self.listbox.insert(tk.END, item)

    def remove_selected(self, event):
        selected_item = self.listbox.curselection()
        if selected_item:
            self.listbox.delete(selected_item)

    
        
```

```python
class HealthBar:
    def __init__(self, master,x):
        self.master = master
        master = Toplevel(x)
        master.title("Barre de vie")
        self.health = 100
        self.health_bar = tk.Canvas(master, bg='white', width=200, height=20)
        self.health_bar.pack()
        self.health_bar.create_rectangle(0, 0, 200, 20, fill='green')
        self.damage_button = tk.Button(master, text="Infliger des dégâts", command=self.take_damage)
        self.damage_button.pack()
        self.last_cube = 240

    def take_damage(self):
        
        if self.health >-10 :
            self.health = self.health - 10         
            self.last_cube = self.last_cube - 20
            if self.health > 66:
                color = 'green'
            elif self.health > 33:
                color = 'yellow'
            else:
                color = 'red'

            self.health_bar.create_rectangle(self.last_cube, 0, 200 -((self.health/100)*200), 20, fill="grey")
            self.health_bar.create_rectangle(0, 0, (self.health/100)*200, 20, fill=color)
        else :
            self.master.destroy()
        

```
